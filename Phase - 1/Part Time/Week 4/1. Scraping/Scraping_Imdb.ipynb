```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc'
```

```python
x = requests.get(url)
content_1 = x.content
content_1
```

```python
soup = BeautifulSoup(content_1, 'html.parser')
soup
```

```python
soup.title.text
```

```python
m_runtime = soup.find(attrs={"class": "runtime"}).text
m_runtime
```

```python
data_divs = soup.find_all("div", {"class" : "lister-item-content" })
print(len(data_divs))
data_divs
```

```python
data_list=[]
data_divs = soup.find_all("div", {"class" : "lister-item-content"})
for data_div in data_divs:
    data_dict = {}
    data_dict['Movie_Name'] = data_div.find(attrs={"class": "lister-item-header"}).text
    data_dict['Year'] = data_div.find(attrs={"class": "lister-item-year text-muted unbold"}).text
    data_dict['Run_time'] = data_div.find(attrs={"class": "runtime"}).text
    data_dict['Votes'] = data_div.find(attrs={"name": "nv"}).text
    data_list.append(data_dict)
```

```python
data_list
```

```python
import pandas as pd

df = pd.DataFrame(data_list)
```

```python
df
```

```python
df.to_csv("Imdb_top_50.csv")
```
