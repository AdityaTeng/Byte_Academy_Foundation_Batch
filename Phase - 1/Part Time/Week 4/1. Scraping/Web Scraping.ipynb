## Scraping:
    Parsing the content from a website and pulling down the content that you want

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc'
response = requests.get(url)
```

```python
print(response, type(response),'\n')
print(response.status_code, type(response.status_code),'\n')
print(response.cookies, type(response.cookies),'\n')
print(response.headers, type(response.headers),'\n')
print(response.content, type(response.content),'\n')
print(response.text, type(response.text),'\n')

r_txt = response.text
```

```python
soup = BeautifulSoup(r_txt, 'html.parser')
rank= soup.find_all('span', class_='lister-item-index unbold text-primary')
movie_name = soup.find_all('h3', class_='lister-item-header')
rating = soup.find_all('div', class_="inline-block ratings-imdb-rating")
certificate = soup.find_all('span', class_="certificate")
genre = soup.find_all('span', class_="genre")
metascore = soup.find_all('div', class_="inline-block ratings-metascore")
description = soup.find_all('p', class_="text-muted")[1::2]
director = soup.find_all('p', class_="")[2:-1]
gross = soup.find_all('p', class_="sort-num_votes-visible")
```
