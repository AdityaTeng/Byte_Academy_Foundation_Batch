```python
from collections import Counter

file1 = open("kinglear.txt", "r+")
```

```python
print(file1.read())
```

```python
with open('kinglear.txt', 'r') as file:
    data = file.read().replace('\n', ' ')

data
```

```python
data = data.lower()
data
```

```python
str_list = data.split()
str_list
```

```python
count = Counter(str_list)
word_count = count.most_common(100)
```

```python
import pandas as pd

df = pd.DataFrame(word_count, columns =['Word', 'Count']) 
```
