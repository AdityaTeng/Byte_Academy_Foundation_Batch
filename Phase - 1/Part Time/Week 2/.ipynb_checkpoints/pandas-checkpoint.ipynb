```python
import pandas as pd

df = pd.read_csv('jobs.csv', index_col ='Unnamed: 0')
```

```python
df
```

```python
len(df)
```

```python
df = df.dropna()
```

```python
df.pivot_table(index='Industry', values = 'title', aggfunc = 'count')
```

```python
def clean_industry(indus):
    indus1 = indus.strip()
    indus2 = indus1.lower()
    if indus2=='infra re & constr':
        indus2 = 'infra, re & constr'
    elif indus2=='other':
        indus2 = 'others'
        
    return indus2

clean_industry('Learning & Education ')

df['Industry'] = df['Industry'].apply(clean_industry)
```

```python
indus = df.pivot_table(index='Industry', values = 'title', aggfunc = 'count', margins = True)
indus
```

```python
indus['Percent %'] = indus['title']/indus[indus.index=='All'].values[0]*100
indus = indus.sort_values(by=['Percent %'], ascending=False)
```
