```python
import pandas as pd

df = pd.read_csv('matches.csv')

def clean_season(season):
    season = str(season)
    season = season.strip()
    return season

df['season']= df['season'].apply(clean_season)
```

```python
season = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

df_1 = df

for i in season:
    df = df_1[df_1['season']==str(i)]
    if len(df)>0:
        batting_first_win = df[df['win_by_runs']!=0]
        batting_second_win = df[df['win_by_runs']==0]
        match_count = len(df)

        bfw_count = len(batting_first_win)
        bsw_count = len(batting_second_win)

        bfw_win_percent = bfw_count/match_count
        bsw_win_percent = bsw_count/match_count

        print('season', i, 'batting first win %:', bfw_win_percent)
        print('season', i, 'batting second win %:', bsw_win_percent,'\n\n')

        continue
```
